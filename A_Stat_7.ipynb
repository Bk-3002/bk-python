```python
import scipy.stats as stats
import numpy as np

sample_mean = 50
sample_std_dev = 5
sample_size = 30  
standard_error = sample_std_dev / np.sqrt(sample_size)
confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

margin_of_error = z_critical * standard_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

confidence_interval
```

```python
import numpy as np
from scipy.stats import chi2

# Observed frequencies (example data)
observed_frequencies = np.array([22, 18, 19, 9, 8, 24])
total = np.sum(observed_frequencies)
expected_frequencies = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20]) * total
chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)
df = len(observed_frequencies) - 1
p_value = chi2.sf(chi_square_statistic, df)

chi_square_statistic, p_value
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Chi-square distribution values
x = np.linspace(0, 30, 1000)
y = chi2.pdf(x, df)

# Plotting the distribution
plt.plot(x, y, label=f'Chi-square distribution with {df} df')
plt.fill_between(x, 0, y, where=(x >= 15), color='grey', alpha=0.5, label='Area for chi-square statistic of 15')
plt.title('Chi-square Distribution')
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.legend()
plt.show()
```

```python
# Given data
total_sample = 1000
coke_preference = 520

# Proportion
p_hat = coke_preference / total_sample

# Standard error
standard_error = np.sqrt(p_hat * (1 - p_hat) / total_sample)

# Confidence interval
confidence_level = 0.99
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_critical * standard_error
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```

```python
# Observed frequencies
observed_frequencies = np.array([45, 55])  # 45 tails and 55 heads

# Expected frequencies for a fair coin
expected_frequencies = np.array([50, 50])

# Chi-square statistic
chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)

# Degrees of freedom
df = len(observed_frequencies) - 1

# P-value
p_value = chi2.sf(chi_square_statistic, df)

chi_square_statistic, p_value
```

```python
# Given data
sample_mean = 72
population_mean = 70
sample_std_dev = 10
sample_size = 30
alpha = 0.05

# Degrees of freedom
df = sample_size - 1

# t-statistic
t_stat = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# p-value (two-tailed)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

t_stat, p_value
```
