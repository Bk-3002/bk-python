Scatter Plot: Used to observe relationships between two numeric variables. It helps in identifying correlations, trends, and patterns in the data.
Line Plot: Useful for visualizing data points over a continuous interval or time period. It shows trends and changes over time.
Bar Plot: Ideal for comparing categorical data. It shows the distribution of a single variable against a categorical variable.
Box Plot: Used to display the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It's useful for detecting outliers and understanding the spread of the data.
Heatmap: Used for representing data in a matrix form where individual values are represented as colors. Itâ€™s helpful for visualizing complex data like correlations between variables or a large matrix of data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")

# Plot box plot for age
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Plot box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple="stack")
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns

# Load the dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to get a matrix form for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Flights Data")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
